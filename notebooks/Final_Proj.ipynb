```python
import numpy as np
def get_flight_status(cancelled, dep_delay):
    flight_status = np.where(cancelled, 'Cancelled', np.where(dep_delay > 0, 'Delayed', 'On Time'))
    return flight_status
```

```python
import pandas as pd
import sklearn
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

print(sklearn.__version__)

le = LabelEncoder()

df = pd.read_csv("Combined_Flights_2021.csv", 
                 usecols=['Airline', 'Origin', 'Dest', 'Cancelled', 'DepTime', 'DepDelay', \
                          'Month', 'DayOfWeek', 'OriginCityName', 'OriginStateName', 'DestCityName', 'DestStateName', 'Distance'], 
                 dtype={  
                'Airline': 'str',
                'Origin': 'str',
                'Dest': 'str',
                'Cancelled': 'bool', # Part of FlightStatus
                'DepTime': 'float64', 
                'DepDelay': 'float64', # Part of FlightStatus
                'Month': 'str',
                'DayOfWeek': 'int64',
                'OriginCityName' : 'str',
                'OriginStateName' : 'str',
                'DestCityName' : 'str',
                'DestStateName' : 'str',
                'Distance': 'float64'
                }, 
                 index_col=None)

out_df = pd.DataFrame()

df['FlightStatus'] = get_flight_status(df['Cancelled'], df['DepDelay'])

le.fit(df['Airline'])
df['Airline'] = le.fit_transform(df['Airline'])
out_df['Airline'] = le.inverse_transform(df['Airline'].unique())
out_df['Airline_Encoded'] = df['Airline'].unique()
pickle.dump(out_df.set_index("Airline")["Airline_Encoded"].to_dict(), open('Airline_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)

out_df = pd.DataFrame()

le.fit(df['Origin'])
df['Origin'] = le.fit_transform(df['Origin'])
out_df['Origin'] = le.inverse_transform(df['Origin'].unique())
out_df['Origin_Encoded'] = df['Origin'].unique()
pickle.dump(out_df.set_index("Origin")["Origin_Encoded"].to_dict(), open('Origin_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['Dest'])
df['Dest'] = le.fit_transform(df['Dest'])
out_df['Dest'] = le.inverse_transform(df['Dest'].unique())
out_df['Dest_Encoded'] = df['Dest'].unique()
pickle.dump(out_df.set_index("Dest")["Dest_Encoded"].to_dict(), open('Dest_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['Month'])
df['Month'] = le.fit_transform(df['Month'])
out_df['Month'] = le.inverse_transform(df['Month'].unique())
out_df['Month_Encoded'] = df['Month'].unique()
pickle.dump(out_df.set_index("Month")["Month_Encoded"].to_dict(), open('Month_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['OriginCityName'])
df['OriginCityName'] = le.fit_transform(df['OriginCityName'])
out_df['OriginCityName'] = le.inverse_transform(df['OriginCityName'])
out_df['OriginCityName_Encoded'] = df['OriginCityName']
pickle.dump(out_df.set_index("OriginCityName")["OriginCityName_Encoded"].to_dict(), open('OriginCityName_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['OriginStateName'])
df['OriginStateName'] = le.fit_transform(df['OriginStateName'])
out_df['OriginStateName'] = le.inverse_transform(df['OriginStateName'].unique())
out_df['OriginStateName_Encoded'] = df['OriginStateName'].unique()
pickle.dump(out_df.set_index("OriginStateName")["OriginStateName_Encoded"].to_dict(), open('OriginStateName_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['DestCityName'])
df['DestCityName'] = le.fit_transform(df['DestCityName'])
out_df['DestCityName'] = le.inverse_transform(df['DestCityName'].unique())
out_df['DestCityName_Encoded'] = df['DestCityName'].unique()
pickle.dump(out_df.set_index("DestCityName")["DestCityName_Encoded"].to_dict(), open('DestCityName_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['DestStateName'])
df['DestStateName'] = le.fit_transform(df['DestStateName'])
out_df['DestStateName'] = le.inverse_transform(df['DestStateName'].unique())
out_df['DestStateName_Encoded'] = df['DestStateName'].unique()
pickle.dump(out_df.set_index("DestStateName")["DestStateName_Encoded"].to_dict(), open('DestStateName_unique.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
out_df = pd.DataFrame()

le.fit(df['FlightStatus'])
df['FlightStatus'] = le.fit_transform(df['FlightStatus'])
print(df['FlightStatus'], le.inverse_transform(df['FlightStatus']))
df = df.reset_index()

df = df.drop(columns='DepDelay')
df = df.drop(columns='Cancelled')
df = df.drop(columns='index')
print(sklearn.__version__)

df
out_df
```

```python
df.fillna(0)
```

```python
%%time
# Decision Tree Clasification
from sklearn import tree
from imblearn.over_sampling import RandomOverSampler
# Train the model
dct = tree.DecisionTreeClassifier(criterion="entropy", max_depth=60)
x = df.drop(columns='FlightStatus')
y = df['FlightStatus']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
dct = dct.fit(X_train, y_train)
X_train.shape

pickle.dump(dct, open('flight_predictor_model.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
```

```python
%%time
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import balanced_accuracy_score
predicts = dct.predict(X_test)
print(f"Accuracy Score: {accuracy_score(y_test, predicts)}")
print(f"F1 Score: {f1_score(y_test, predicts, average='weighted')}")
print(f"Balanced Accuracy Score: {balanced_accuracy_score(y_test, predicts)}")
X_test.shape
```

```python
%%time
# Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp = imp.fit(df)
x =  pd.DataFrame(imp.transform(df), columns=df.columns, index=df.index).drop(columns='FlightStatus')
y =  pd.DataFrame(imp.transform(df), columns=df.columns, index=df.index)['FlightStatus']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42, stratify=y)
gs = GaussianNB()
gs = gs.fit(X_train, y_train)
X_train.shape
```

```python
%%time
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
predicts = gs.predict(X_test)
print(f"Accuracy Score: {accuracy_score(y_test, predicts)}")
print(f"F1 Score: {f1_score(y_test, predicts, average='weighted')}")
print(f"Balanced Accuracy Score: {balanced_accuracy_score(y_test, predicts)}")
X_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

# Hyperparameter Tuning with GridSearchCV

parameters = {"max_depth" : [2, 3, 5, 8, 13, 21, 34, 55, 89, 95], "criterion" : ("gini", "entropy"), \
              "max_features" : ("sqrt", "log2"), "min_samples_split" : [2, 4, 8, 16, 24, 32, 40]}
# Train the model
dct = GridSearchCV(tree.DecisionTreeClassifier(), parameters)

x = df.drop(columns='FlightStatus')
y = df['FlightStatus']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)

with parallel_backend('multiprocessing', n_jobs=4):
    dct = dct.fit(X_train, y_train)

X_train.shape

dct.best_estimator_
dct.best_score_
dct.best_params_

dfa = pd.DataFrame(dct.cv_results_)
dfa.to_csv("cv_results.csv", index = False)
# pickle.dump(dct, open('flight_predictor_model.pickle', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
```
